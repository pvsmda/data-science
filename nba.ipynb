```python
# Import dependencies
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
players_url = "https://raw.githubusercontent.com/will-lucena/Data-science-class-project/master/Players.csv"
player_data_url = "https://raw.githubusercontent.com/will-lucena/Data-science-class-project/master/player_data.csv"
season_stats_url = "https://raw.githubusercontent.com/will-lucena/Data-science-class-project/master/Seasons_Stats.csv"
players_list_data = pd.read_csv(players_url)
players_data = pd.read_csv(player_data_url)
season_data = pd.read_csv(season_stats_url)
```

```python
# print(players_list_data.head())
# print(players_list_data.info())
players_list_data[players_list_data['Player'] == 'Kobe Bryant']
```

```python
# print(season_data.head())
# print(season_data.info())
season_data[season_data['Player'] == 'Kobe Bryant']
```

```python
groupDf = season_data.groupby(['Player'])
df = groupDf.head().values
print(df.size)
print(df)
```

```python
df_2grp = season_data.groupby(['Player', 'Year'])
target_year = 2010
array = df_2grp.head()[df_2grp.head()['Year'] == target_year]
print(array.values.size)
print(array.values)
```

```python
index = 0
other_index = 2
first_player = array.values[index]
second_player = array.values[other_index]
print(first_player[2])
print(second_player[2])
first_plot_data = groupDf.get_group(first_player[2])
second_plot_data = groupDf.get_group(second_player[2])
```

```python
plt.xlabel('Years') 
plt.ylabel('Total points')
y = first_plot_data['PTS']
x = first_plot_data['Year']
plt.plot(x, y)
plt.show()
second_plot_data = second_plot_data.drop(18229, axis=0)
y = second_plot_data['PTS']
x = second_plot_data['Year']
plt.plot(x, y)
plt.show()
```
